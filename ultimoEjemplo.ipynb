```python
x = [18, 25, 35, 33, 19, 20, 22, 32, 20, 21, 22, 24, 27, 28, 19] #Edad,
y = [90, 88, 83, 95, 88, 62, 90, 96, 78, 70, 65, 88, 93, 99, 60] #Puntaje

""" 
    Graficar la disperción de los puntajes según la edad 
    Obtener la recta de la regresión lineal
    Predecir el puntaje de un alumno de edad 29 y 31
    Analis sobre la relación de los datos
"""
import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x,y)
print(r*r) # 0% - 100%

def funRecta(x):
    return slope * x + intercept

modelo = list(map(funRecta, x))

plt.scatter(x,y,marker="^")
plt.plot(x, modelo, "r")
plt.title("Edad vs Puntaje", color="blue", style="italic")
plt.show()

edad29 = funRecta(29)
print(edad29)

edad31 = funRecta(31)
print(edad31)


print("La relación de los datos tiene un índice de confiabilidad de: " , round((r*r),4)*100, "%") #Para que quede en porcentaje
```

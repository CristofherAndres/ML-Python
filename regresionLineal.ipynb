```python
""" Gráfico de dispersión """

x = [1, 5, 12, 4, 6, 11, 12, 1, 2, 15] #Sueldos
y = [0, 3, 6, 2, 3, 5, 4, 1, 1, 3] #N de casas

import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x,y)

def funRecta(x):
    return slope * x + intercept

modelo = list(map(funRecta, x))

plt.scatter(x,y,marker="^")
plt.plot(x, modelo, "r")
plt.title("Titulo del grafico", color="blue", style="italic")
plt.show()

```

```python
""" Usar datos de distribución normal """

import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(5.0, 0.5, 1000)
y = np.random.normal(2, 1, 1000)
plt.scatter(x,y)
plt.show()
```

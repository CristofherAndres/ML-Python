```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] # Año del auto
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # Velocidad que tiene el auto

import matplotlib.pyplot as plt
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x,y)
print(r*r)

def funRecta(x):
    return slope * x + intercept

modelo = list(map(funRecta, x))

plt.scatter(x,y,marker="^")
plt.plot(x, modelo, "r")
plt.title("Años vs Velocidad", color="blue", style="italic")
plt.show()

velocidad10 = funRecta(10)
print(velocidad10)

velocidad3 = funRecta(3)
print(velocidad3)

""" GRaficar un punto en recta y con una etiqueta """


```

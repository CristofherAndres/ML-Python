```python
""" Crear una matriz de 300 datos que sus valores esten entre 0 y 10
    sea una distribución uniforme """

import numpy
import matplotlib.pyplot as plt

datos = numpy.random.uniform(0.0, 10.0, 300)
print(datos)

plt.hist(datos, 10)
plt.show()

```

```python
""" Ver el promedio de la muestra """

print(numpy.mean(datos))
```

```python
""" Distribución normal de datos  """

import numpy
import matplotlib.pyplot as plt

datos = numpy.random.normal(4, 1.9, 300)
print(datos)

plt.hist(datos, 10)
plt.show()

```
